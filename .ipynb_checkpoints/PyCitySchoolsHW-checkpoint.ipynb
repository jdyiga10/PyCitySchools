```python
# Define the spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create the "Spending Ranges (Per Student)" column using pd.cut()
school_data["Spending Ranges (Per Student)"] = pd.cut(school_data["budget"] / school_data["size"], spending_bins, labels=labels)

# Display the DataFrame to see the new column
print(school_data.head())
```

```python
# Scores by School Spending
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# create a pd.cut
schools_summary['Spending Ranges (Per Student)'] = pd.cut(schools_summary['Total School Budget'] / schools_summary['Total Students'], bins=spending_bins, labels=spending_labels)

# mean scores per spending range
spending_math_scores = schools_summary.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = schools_summary.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = (schools_summary["Average Math Score"] >= 70).groupby(schools_summary["Spending Ranges (Per Student)"]).mean() * 100
spending_passing_reading = (schools_summary["Average Reading Score"] >= 70).groupby(schools_summary["Spending Ranges (Per Student)"]).mean() * 100
overall_passing_spending = ((schools_summary["Average Math Score"] >= 70) & (schools_summary["Average Reading Score"] >= 70)).groupby(schools_summary["Spending Ranges (Per Student)"]).mean() * 100
```

```python
# spending summary DataFrame
spending_summary = pd.DataFrame({'Average Math Score': spending_math_scores,
    'Average Reading Score': spending_reading_scores,
    '% Passing Math': spending_passing_math,
    '% Passing Reading': spending_passing_reading,
    '% Overall Passing': overall_passing_spending})
spending_summary
```

```python
# Scores by School Size
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# create a pd.cut
schools_summary['School Size'] = pd.cut(schools_summary['Total Students'], bins=size_bins, labels=size_labels)

# mean calculations for scores by school size
size_math_scores = schools_summary.groupby('School Size')['Average Math Score'].mean()
size_reading_scores = schools_summary.groupby('School Size')['Average Reading Score'].mean()
size_passing_math = schools_summary.groupby('School Size')['% Passing Math'].mean()
size_passing_reading = schools_summary.groupby('School Size')['% Passing Reading'].mean()
overall_passing_size = schools_summary.groupby('School Size')['% Overall Passing'].mean()
```

```python
# size summary
size_summary = pd.DataFrame({
    'Average Math Score': size_math_scores,
    'Average Reading Score': size_reading_scores,
    '% Passing Math': size_passing_math,
    '% Passing Reading': size_passing_reading,
    '% Overall Passing': overall_passing_size
})
size_summary
```

```python
# Scores by School Type
type_math_scores = schools_summary.groupby('School Type')['Average Math Score'].mean()
type_reading_scores = schools_summary.groupby('School Type')['Average Reading Score'].mean()
type_passing_math = schools_summary.groupby('School Type')['% Passing Math'].mean()
type_passing_reading = schools_summary.groupby('School Type')['% Passing Reading'].mean()
overall_passing_type = schools_summary.groupby('School Type')['% Overall Passing'].mean()
```

```python
# school type summary
type_summary = pd.DataFrame({
    'Average Math Score': type_math_scores,
    'Average Reading Score': type_reading_scores,
    '% Passing Math': type_passing_math,
    '% Passing Reading': type_passing_reading,
    '% Overall Passing': overall_passing_type})
type_summary
```
